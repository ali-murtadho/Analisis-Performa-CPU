```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
dtr = pd.read_excel("cpu_training.xlsx")
dtr
```

```python
dts = pd.read_excel("cpu_testing.xlsx")
dts
```

```python
ds = pd.read_excel("cpu.xls")
ds
```

```python
x = ds.iloc[:,0:6].values
y = ds['PERFORMANCE'].values
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.08)
```

```python
print("Jumlah Data Training = ", xtrain.shape)
print("Jumlah Data Testing = ", xtest.shape)
```

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential()
```

```python
model.add(tf.keras.layers.Dense(units=6, activation='relu'))
model.add(tf.keras.layers.Dense(units=9, activation='relu'))
model.add(tf.keras.layers.Dense(units=1))
model.compile(loss='mean_absolute_error', optimizer=tf.keras.optimizers.Adam(0.001))
model.fit(xtrain, ytrain, epochs=800, batch_size=128)
```

```python
pred = model.predict(xtest)
pred
```

```python
plt.plot(ytest, pred, 'b', "Hasil Prediksi")
plt.figure()
```

```python
model.evaluate(xtest, ytest)
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(ytest, pred)
mae
```

```python
a = range(len(ytest))

plt.scatter(a,pred, color = 'blue')
plt.scatter(a,ytest, color='red')
plt.title("Hasil Prediksi")
plt.xlabel("Data Testing")
plt.ylabel("Data Aktual")
plt.show()
```
